1. Each component in the cycle is analyzed as a control volume at steady state. The control volumes are shown on the

```python

from seuif97 import *

def fixeachoftheprincipalstates():
     p = []
     t = []
     h = []
     s = []
     v = []
     x = []

     numNodes = 14

     for node in range(numNodes):
        p.append(None)
        t.append(None)
        h.append(None)
        s.append(None)
        v.append(None)
        x.append(None)
     #Given
     p[1]=8
     p[2]=2
     p[3]=0.7
     p[4]=0.7
     p[5]=0.3
     p[6]=0.008
     p[7]=0.008
     p[8]=0.3
     p[9]=0.3
     p[10]=8
     p[11]=8
     p[12]=2
     t[1]=480;t[4]=440;     

     #**step**
     #from the picture we knows that entropy s3=s2=s1,Also s6=s5=s4
    
     #Node1
     h[1]=pt2h(p[1],t[1])          
     s[1]=pt2s(p[1],t[1]) 
        
     #Node2
     s[2]=s[1]
     h[2]=ps2h(p[2],s[2])
    
     #Node3
     s[3]=s[1]
     h[3]=ps2h(p[3],s[3])
        
     #Node4
     h[4]=pt2h(p[4],t[4])
     s[4]=pt2s(p[4],t[4])
    
     #Node5
     s[5]=s[4]
     h[5]=ps2h(p[5],s[5])
     
     #state 7 is the satura#hfg is the saturated entropy at t=4205ted liquid at pressure p7
     #state 9 is the saturated liquid at pressure p9
     #state 12 is the saturated liquid at pressure p12
     #Node6、7
     s[6]=s[4]
     h[7]=px2h(p[7],0)
     v[7]=px2v(p[7],0)*1000
     h[6]=ps2h(p[6],s[6])    
     #Node8
     h[8]=h[7]+v[7]*(p[8]-p[7]) 
    
     #Node9
     h[9]=px2h(p[9],0)
     v[9]=px2v(p[9],0)*1000 
        
     #Node10
     h[10]=h[9]+v[9]*(p[10]-p[9])
 
     #Node11
     h[11]=pt2h(p[11],205)
     #Node12
     h[12]=px2h(p[12],0)
    
     #Node13
     h[13]=h[12]
     return(p,t,h,s,v,x)
```

$$y'=\frac{h_{11}-h_{10}}{h_{2}-h_{12}}$$

$$y''=\frac{(1-y')*h_8+y'*h_{13}-h_9}{h_8-h_5}$$

```python

def yy():
     y=[]
     numNodes = 3

     for node in range(numNodes):
        y.append(None)
    
     y[1] = (h[11] - h[10]) / (h[2] - h[12])
     y[2] = ((1 - y[1]) * h[8] + y[1]*h[13] - h[9]) / (h[8] - h[5])
     return(y)
```

* 然后，需要计算出郎肯循环中各个机器的做工量和整个郎肯循环总体的吸热量，当然是在单位进入第一级汽轮机蒸汽量下计算得出的。  
第一级汽轮机：$\frac{W_{t1}}{m_1}=h_1-h_2+(1-y')(h_2-h_3)$  
第二级汽轮机：$\frac{W_{t2}}{m_1}=(1-y')(h_4-h_5)+(1-y'-y'')(h_5-h_6)$  
第一个泵：$\frac{W_{p1}}{m_1}=(1-y'-y'')(h_8-h_7)$  
第二个泵：$\frac{W_{p2}}{m_1}=h_10-h_9$  
循环总的吸热量：$\frac{Q_{in}}{m_1}=h_1-h_11+(1-y)(h_4-h_3)$  
* 郎肯循环的热效率:$$\eta=\frac{\frac{W_{t1}}{m_1}+\frac{W_{t2}}{m_1}-\frac{W_{p1}}{m_1}-\frac{W_{p2}}{m_1}}{\frac{Q_{in}}{m_1}}$$
* 流入第一级汽轮机的蒸汽流:$$m_1=\frac{W_{cycle}}{\frac{W_{t1}}{m_1}+\frac{W_{t2}}{m_1}-\frac{W_{p1}}{m_1}-\frac{W_{p2}}{m_1}}$$
  这些数据处理的结果全都保存在字典中返回。

```python
#Answer
from seuif97 import *

def Determineforthecycle(h,y,Wcycledot):
    results={}
    #(a)
    results['wtdot1']=h[1]-y[1]*h[2]-(1-y[1])*h[3]
    results['wtdot2']=(1-y[1])*h[4]-y[2]*h[5]-(1-y[1]-y[2])*h[6]
    results['wtdot']=results['wtdot1']+results['wtdot2']
    results['wpdot']=(1-y[1]-y[2])*(h[8]-h[7])+h[10]-h[9]
    results['qindot']=(h[1]-h[11])+(1-y[1])*(h[4]-h[3])
    results['eta'] = ( results['wtdot1']+results['wtdot2']-results['wpdot'])/results['qindot']

    #(b)
    results['mcwdot'] = (Wcycledot*3600*10**3)/(results['wtdot1']+results['wtdot2']-results['wpdot'])
    return results


#Results
Wcycledot = 100.0
p,t,h,s,v,x=fixeachoftheprincipalstates() 
y=yy() 
cycleResults=Determineforthecycle(h,y,Wcycledot)
print('(a) The thermal efficiency for the cycle is %.3f '%(cycleResults['eta']*100),'%')
print('(b) The mass flow rate of the condenser cooling water is %.3f '%(cycleResults['mcwdot']),'kg/h')
```

### 小结(我犯过的错误）

 （1）程序块不是以“{”和“}”括起来来定义的，而是依据程序的缩进来表示程序的块和代码的层次。 
    
 （2）程序中没有类似main函数的主程序，一个文件中的所有内容（代码）按照行从上到下从左到右执行。在文件中有函数，函数定义部分等到调用才执行。
 
 （3）函数定义中没有返回值及返回值类型的说明，只有函数名和参数列表（参数也没有类型，同一函数在同一程序中可接收不同类型的参数）。如果没有返回值则默认返回NONE，如果有显示的return语句，则返回被return的值。（在同一函数，多个return可以返回不同类型的或不返回）
 
 （4）print 函数：打印内容到控制台，可以是数值、字符串、序列等各种类型。默认换行可以格式化字符串，使用%d（%o、%x）、%f、%s等。
 
 （5）函数定义没有返回值的说明，但是可以使用return语句返回一个值，如果在函数中有多个return语句，那么返回值类型可以不同，如果没有return语句，默认是返回None。函数的参数列表也只有形参名称，而没有形参的类型，具体的类型根据运行时动态确定，在同一程序中，可以给同一函数传递不同类型的参数来调用这个函数。注意函数定义时后面跟“:”。函数调用的形式为functionname(arg1, arg2, …… ,argn)。
 
 （6）列表 特点：列表是可变的，可以删除、排序、追加（列表中的列表也不会失去它的身份），这里需要用列表定义每个节点
 
 （7）元组 特点：元组是不可变的。 元组和列表十分类似，只不过元组和字符串一样是不可变的，即你不能修改元组。每个变量都和上面一个变量有关联而不是一开始知道，因此此题不应用元组
