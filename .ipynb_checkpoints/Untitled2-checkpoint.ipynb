## 2.程序说明
###2.1我为什么要用元组
元组是不可变的， 而列表是可变的。元祖的顺序不可改变且不能删除。故在这种不可改变的热力循环顺序中，用元组更为合适,且列表不可以当作字典的key， 而元组可以
###2.我为什么要用字典
Dict：key-value形式存放、使用key访问数据（hash），查找速度快，不会随着元素增多而变慢，时间复杂度O(1)；默认无序（需要排序可以使用collections.OrderDict）；dict可以嵌套list,把无序的数据调整为有序的数据，可以快速访问特征，大大减少查找量
## 3.Asumptions
1. Each component in the cycle is analyzed as a control volume at steady state. The control volumes are shown on the
accompanying sketch by dashed lines.
2. There is no stray heat transfer from any component to its surroundings.
3. The working fluid undergoes internally reversible processes as it passes through the turbines, pumps, steam generator,
reheater, and condenser.
4. The expansion through the trap is a throttling process.
5. Kinetic and potential energy effects are negligible.
6. Condensate exits the closed heater as a saturated liquid at 2 MPa. Feedwater exits the open heater as a saturated liquid at
0.3 MPa. Condensate exits the condenser as a saturated liquid.

```python

from seuif97 import *

def fixeachoftheprincipalstates():
     p = []
     t = []
     h = []
     s = []
     v = []
     x = []

     numNodes = 14

     for node in range(numNodes):
        p.append(None)
        t.append(None)
        h.append(None)
        s.append(None)
        v.append(None)
        x.append(None)
     #Given
     p[1]=8
     p[2]=2
     p[3]=0.7
     p[4]=0.7
     p[5]=0.3
     p[6]=0.008
     p[7]=0.008
     p[8]=0.3
     p[9]=0.3
     p[10]=8
     p[11]=8
     p[12]=2
     t[1]=480;t[4]=440;     

     #**step**
     #from the picture we knows that entropy s3=s2=s1,Also s6=s5=s4
    
     #Node1
     h[1]=pt2h(p[1],t[1])          
     s[1]=pt2s(p[1],t[1]) 
        
     #Node2
     s[2]=s[1]
     h[2]=ps2h(p[2],s[2])
    
     #Node3
     s[3]=s[1]
     h[3]=ps2h(p[3],s[3])
        
     #Node4
     h[4]=pt2h(p[4],t[4])
     s[4]=pt2s(p[4],t[4])
    
     #Node5
     s[5]=s[4]
     h[5]=ps2h(p[5],s[5])
     #hfg is the saturated entropy at t=4205
     #state 7 is the saturated liquid at pressure p7
     #state 9 is the saturated liquid at pressure p9
     #state 12 is the saturated liquid at pressure p12
     hfg=tx2h(205,1)
     #Node6、7
     s[6]=s[4]
     x[6]=ps2x(p[6],s[6])
     h[7]=px2h(p[7],0)
     v[7]=px2v(p[7],0)*1000
     h[6]=h[7]+x[6]*hfg
    
     #Node8
     h[8]=h[7]+v[7]*(p[8]-p[7]) 
    
     #Node9
     h[9]=px2h(p[9],0)
     v[9]=px2v(p[9],0)*1000 
        
     #Node10
     h[10]=h[9]+v[9]*(p[10]-p[9])
    
     #Assump state f is saturated at t=205
     hf=tx2h(205,0)
     vf=tx2v(205,0)*1000
     pf=tx2p(205,0)
        
     #Node11
     h[11]=hf+vf*(p[11]-pf)
    
     #Node12
     h[12]=px2h(p[12],0)
    
     #Node13
     h[13]=h[12]
     return(p,t,h,s,v,x)
```

$$y'=\frac{h_{11}-h_{10}}{h_{2}-h_{12}}$$

$$y''=\frac{(1-y')*h_8+y'*h_{13}-h_9}{h_8-h_5}$$

```python
from seuif97 import *

def yy():
     y=[]
     numNodes = 3

     for node in range(numNodes):
        y.append(None)
    
     y[1] = (h[11] - h[10]) / (h[2] - h[12])
     y[2] = ((1 - y[1]) * h[8] + y[1]*h[13] - h[9]) / (h[8] - h[5])
     return(y)
```

```python
#Answer
from seuif97 import *

def Determineforthecycle(h,y,Wcycledot):
    results={}
    #(a)
    results['wtdot1']=h[1]-y[1]*h[2]-(1-y[1])*h[3]
    results['wtdot2']=(1-y[1])*h[4]-y[2]*h[5]-(1-y[1]-y[2])*h[6]
    results['wtdot']=results['wtdot1']+results['wtdot2']
    results['wpdot']=(1-y[1]-y[2])*(h[8]-h[7])+h[10]-h[9]
    results['qindot']=(h[1]-h[11])+(1-y[1])*(h[4]-h[3])
    results['eta'] = ( results['wtdot1']+results['wtdot2']-results['wpdot'])/results['qindot']

    #(b)
    results['mcwdot'] = (Wcycledot*3600*10**3)/(results['wtdot1']+results['wtdot2']-results['wpdot'])
    return results


#Results
Wcycledot = 100.0
p,t,h,s,v,x=fixeachoftheprincipalstates() 
y=yy() 
cycleResults=Determineforthecycle(h,y,Wcycledot)
print('(a) The thermal efficiency for the cycle is %.3f '%(cycleResults['eta']*100),'%')
print('(b) The mass flow rate of the condenser cooling water is %.3f '%(cycleResults['mcwdot']),'kg/h')
```
