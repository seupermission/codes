```python
# **软件工程基础及实践作业二**

### 03015414 王瑄

## **1.问题描述**

两台给水加热器(一台封闭式给水加热器和一台开式给水加热器)的再热-回热循环。

* 蒸汽在8.0MPa，480℃进入第一台汽轮机并膨胀到0.7MPa。

* 蒸汽在进入第二台汽轮机前被加热到440°C，并在此膨胀到0.008 MPa的冷凝器压力。

* 在2MPa下从第一个汽轮机中抽取蒸汽并将其送入封闭式给水加热器。

* 给水离开闭式加热器在205℃和8.0MPa，冷凝物以2MPa作为饱和液体离开。

* 冷凝水进入开式给水加热器中。

* 以0.3MPa从第二个汽轮机抽取的蒸汽也被送入开式给水加热器，该加热器在0.3MPa下运行。离开开式给水加热器的物流是0.3MPa的饱和液体。

* 该循环的净功率输出为100MW。

* 不会有任何组件散布到周围环境的热量传递。

* 如果工作流体在穿过涡轮机，泵，蒸汽发生器，再热器和冷凝器时没有经历不可逆性，

* 求解

  * 已知：再热再生蒸气动力循环以蒸汽作为工作流体进行操作。指定工作压力和温度，并给出净功率输出。

  * 求出：确定进入第一台涡轮机的热效率和质量流量，单位为kg / h。
    ![](./img/fig86.jpg)
    
## **2.设计思路**

   带回热和再热的朗肯循环可以分为几部分：锅炉加热，汽轮机做工，冷凝，泵做工。其中还有回热和再热循环。使用列表表示各个点的状态。通过图中数据和各点之间的关系，我们可以调用seuif97中的函数，求出各个点的其余状态量。再根据公式求出热效率和流量。
    
## **3.模块说明**

   具体流程如图：
    
   ![朗肯循环](./img/朗肯循环.bmp)
    
    
    
## **4.小结**

   （1）本次练习中，我们使用列表表示每个节点的状态。在编程过程中，我学习到元组、列表、字典等数据结构。列表与元组再定义和操作上有很多相似之处，可以相互转换，但也有很多不同。列表中的元素可以更改，但元组的元素一旦定义就不能改变。因此元组可以做字典键，而列表不可以。由于本题中各节点都与其它节点有关，也就是说元素需要改变，因此不可用元组。
   （2）在函数方面，不需要大括号来表示作用域，需要通过缩进。
   （3）在计算y1，y2时，需要定义一个y函数，才能在Determineforthecycle函数中使用
    

  
```

```python
from seuif97 import *

def fixeachoftheprincipalstates():
    p=[]
    t=[]
    h=[]
    s=[]
    v=[]
    x=[]

    numNodes=13

    for node in range(numNodes):
        p.append(None)
        t.append(None)
        h.append(None)
        s.append(None)
        v.append(None)
        x.append(None)
 
    #state 1  
    p[0]=8.0
    t[0]=480.0
    h[0]=pt2h(p[0],t[0])
    s[0]=pt2s(p[0],t[0])
    
    #state 2
    p[1]=2.0
    s[1]=s[0]
    h[1]=ps2h(p[1],s[1])
    
    #state 3
    p[2]=0.7
    s[2]=s[0]
    h[2]=ps2h(p[2],s[2])
    
    #state 4
    t[3]=440.0
    p[3]=p[2]
    s[3]=pt2s(p[3],t[3])
    h[3]=pt2h(p[3],t[3])
    
    #state 5
    p[4]=0.3
    s[4]=s[3]
    h[4]=ps2h(p[4],s[4])
    
    #state 6
    p[5]=0.008
    s[5]=s[3]
    sf1=px2s(p[5],0)
    sg1=px2s(p[5],1)
    x[5]=(s[5]-sf1)/(sg1-sf1)
    
    hf1=px2h(p[5],0)
    hg1=px2h(p[5],1)
    h[5]=hf1+(hg1-hf1)*x[5]
    
    #state 7
    p[6]=p[5]
    s[6]=sf1
    h[6]=hf1
    v[6]=px2v(p[6],0)
    
    #state 8
    s[7]=s[6]
    p[7]=0.3
    h[7]=h[6] + v[6]*(p[7]-p[6])*10**6*10**-3 
    
    #state 9
    p[8]=p[7]
    s[8]=px2s(p[8],0)
    h[8]=px2h(p[8],0)
    v[8]=px2v(p[8],0)
    
    #state 10
    s[9]=s[8]
    p[9]=8.0
    h[9]=h[8] + v[8]*(p[9]-p[8])*10**6*10**-3 
    
    #state 11
    t[10]=205.0
    p[10]=p[0]
    h[10]=pt2h(p[10],t[10])

    #state 12
    p[11]=2.0
    h[11]=px2h(p[11],0)
    s[11]=px2s(p[11],0)
    v[11]=px2v(p[11],0)
  
    
    #state 13
    p[12]=0.3
    h[12]=h[11]
    
    return (p,t,h,s,v,x)
    
```

 
 $$y'=\frac{h_{11}-h_{10}}{h_{2}-h_{12}}$$

 $$y''=\frac{h_{9}-h_{8}+(h_{8}-h_{13})*y'}{h_8-h_5}$$

```python
def y():
    y=[]
    numNodes = 2

    for node in range(numNodes):
         y.append(None)
    #the steam extracted from the first turbin y1
    #y1*h2+h10=h11+h12*y1
    y[0]=(h[10]-h[9])/(h[1]-h[11])
    
    #the steam extracted from the second turbin y2
    #h5*y2+h8*(1-y1-y2)+y1*h13=h9
    y[1]=(h[8]-h[7]+(h[7]-h[12])*y[0])/(h[4]-h[7])
    return (y)
```

效率
    $$Ψ=\frac{Wt-Wp1-Wp2}{Qin}$$

流量
    $$Qm=\frac{Wc}{Wt-Wp1-Wp2}$$

```python

def Determineforthecycle(h,y,Wcycle): 
    
    results={}
    #generator qin
    results['qin']=h[0]-h[10]
    
    #pump 1 
    results['wp1']=(h[7]-h[6])*(1-y[0]-y[1])
    #pump 2
    results['wp2']=h[9]-h[8]
   
    
    #turbin
    results['wt']=(h[0]-h[1])+(h[1]-h[2])*(1-y[0])+(1-y[0])*(h[3]-h[4])+(1-y[0]-y[1])*(h[4]-h[5])
    
    #The heat transfer to the working fluid  as it passes through the boiler
    results['qin']=h[0]-h[10]+(1-y[0])*(h[3]-h[2])
    
    #the thermal efficiency
    results['eta'] = ( results['wt'] -results['wp1']-results['wp2'])/ results['qin'] 
                                                       
    #the mass flow rate entering the first turbine
    results['m'] = (Wcycle*10**3*3600)/( results['wt'] -results['wp1']-results['wp2'])
    
    return (results)


Wcycle = 100.00 
p,t,h,s,v,x = fixeachoftheprincipalstates() 
y=y()
cycleResults = Determineforthecycle(h,y,Wcycle)

print('(a) The thermal efficiency for the cycle is %.3f '%(cycleResults['eta']*100),'%')
print('(b) The mass flow rate entering the first turbine is %.2f' %cycleResults['m'],'kg/h.')
```
