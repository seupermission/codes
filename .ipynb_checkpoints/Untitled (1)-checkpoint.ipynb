```python
from seuif97 import *

def fixeachoftheprincipalstates():
    p=[]
    t=[]
    h=[]
    s=[]
    v=[]
    x=[]

    numNodes=13

    for node in range(numNodes):
        p.append(None)
        t.append(None)
        h.append(None)
        s.append(None)
        v.append(None)
        x.append(None)
 
    #state 1  
    p[0]=8.0
    t[0]=480.0
    h[0]=px2h(p[0],1)
    s[0]=px2s(p[0],1)
    
    #state 2
    p[1]=2.0
    s[1]=s[0]
    
    #state 3
    p[2]=0.7
    s[2]=s[0]
    sf=px2s(p[2],0)
    sg=px2s(p[2],1)
    x[2]=(s[2]-sf)/(sg-sf)
    
    hf=px2h(p[2],0)
    hg=px2h(p[2],1)
    h[2]=hf+(hg-hf)*x[2]
    
    #state 4
    t[3]=440.0
    s[3]=tx2s(t[3],1)
    h[3]=tx2h(t[3],1)
    
    #state 5
    p[4]=0.3
    s[4]=s[3]
    h[4]=px2h(p[4],1)
    
    #state 6
    p[5]=0.008
    s[5]=s[3]
    sf1=px2s(p[5],0)
    sg1=px2s(p[5],1)
    x[5]=(s[5]-sf1)/(sg1-sf1)
    
    hf1=px2h(p[5],0)
    hg1=px2h(p[5],1)
    h[5]=hf1+(hg1-hf1)*x[5]
    
    #state 7
    p[6]=p[5]
    s[6]=sf1
    h[6]=hf1
    v[6]=px2v(p[6],0)
    
    #state 8
    s[7]=s[6]
    p[7]=0.3
    h[7]=h[6] + v[6]*(p[7]-p[6])*10**6*10**-3 
    
    #state 9
    p[8]=p[7]
    s[8]=px2s(p[8],0)
    h[8]=px2h(p[8],0)
    v[8]=px2v(p[8],0)
    
    #state 10
    s[9]=s[8]
    p[9]=8.0
    h[9]=h[8] + v[8]*(p[9]-p[8])*10**6*10**-3 
    
    #state 11
    t[10]=205.0
    p[10]=p[0]
    hf2=px2h(t[10],0)
    vf=px2v(t[10],0)
    pf=tx2p(t[10],0)
    h[10]=hf2+vf*(p[10]-pf)

    #state 12
    p[11]=2.0
    h[11]=px2h(p[11],0)
    s[11]=px2s(p[11],0)
    v[11]=px2v(p[11],0)
  
    
    #state 13
    p[12]=0.3
    h[12]=h[11]+v[11]*(p[12]-p[11])
    
    return (p,t,h,s,v,x)
    
    
    
        
```

```python
from seuif97 import *

def Determineforthecycle(h,Wcycle):
    
    results={}
    y={}
    
     
    #the steam extracted from the first turbin y1
    #y1*h2+(1-y1)*h10=h11
    y[1]=(h[10]-h[9])/(h[1]-h[9])
    
    #the steam extracted from the second turbin y2
    #h5*y2+h8*(1-y1-y2)+y1*h13=(1-y1)*h9
    y[2]=(h[8]-h[7]+(h[7]-h[8]-h[12])*y[1])/(h[4]-h[7])
    #generator qin
    results['qin']=h[0]-h[10]
    
    #pump 1 
    results['wp1']=(h[7]-h[6])*(1-y[1]-y[2])
    #pump 2
    results['wp2']=(h[9]-h[8])*(1-y[1])
   
    
    #turbin
    results['wt']=(h[0]-h[2])+(1-y[1])*(h[3]-h[4])+(1-y[1]-y2)*(h[4]-h[5])
    
    #The heat transfer to the working fluid  as it passes through the boiler
    results['qin']=h[0]-h[10]+(1-y[1])*(h[3]-h[2])
    
    #the thermal efficiency
    results['eta'] = ( results['wt'] -results['wp1']-results['wp2'])/ results['qin'] 
                                                       
    #the mass flow rate entering the first turbine
    results['m'] = (Wcycle*10**3*3600)/( results['wt'] -results['wp1']-results['wp2'])
    
    return results
```

```python
Wcycle = 100.00 
p,t,h,s,v,x = fixeachoftheprincipalstates()  
cycleResults = Determineforthecycle(h,Wcycle)

print('(a) The thermal efficiency for the cycle is %.3f '%(cycleResults['eta']*100),'%')
print('(b) The mass flow rate entering the first turbine is %.2f' %cycleResults['m'],'kg/h.')
```
