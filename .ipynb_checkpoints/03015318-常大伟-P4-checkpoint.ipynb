## 2.具体实现

### (1) class Group

#### 1) class Group 对一组数据的X和Y进行处理。包括以下函数：


* def __init__(self, X, Y)：构造函数。
 

* def indicators(self)：计算X和Y的均值(mean)、标准差(stdev)和方差(variance)，储存在字典中。
 
 
* def rSquared(self,measured, predicted)：计算线性回归的相关指数$R^2$。
 
 
* def LinearRegression(self):对数据进行处理和拟合，有线性拟合和多项式拟合（包括二次和三次），储存拟合方程的系数等数据。
 
 
* def export(self,j=5): 将数据进行描述，画出数据散点图和回归曲线，输出统计和回归结果。其中参数j在拟合最后一组数据时起作用，调用最后一组数据的export函数，使j=0<5，从而只画linear fit（线性回归）的图。

```python
import numpy as np
import matplotlib.pyplot as plt
from  statistics import mean,stdev,variance

class Group:

    def __init__(self,X,Y,j):
        self.X = np.array(X)
        self.Y = np.array(Y)
        self.X1 = X
        self.Y1 = Y
        self.j = j #统计建立类的个数
        self.r = np.corrcoef(self.X1,self.Y1)[0, 1]  #Pearson correlation 
        

    def indicators(self):
        x={}
        y={}
       
        x['x_avg'] = mean(self.X)
        x['x_std'] = stdev(self.X)
        x['x_pstd'] = stdev(self.X)*10/11
        x['x_var'] = variance(self.X)
        x['x_pvar'] = variance(self.X)*10/11
        y['y_avg'] = mean(self.Y)
        y['y_std'] = stdev(self.Y)
        y['y_pstd'] = stdev(self.Y)*10/11
        y['y_var'] = variance(self.Y)
        y['y_pvar'] = variance(self.Y)*10/11
        return x,y,self.r

    
    def rSquared(self,measured, predicted):
        # RSS: residual sum of squares
        estimateError = ((predicted - measured)**2).sum()
        meanOfMeasured = measured.sum()/float(len(measured))
        
        # TSS: total sum of squares
        variability = ((measured - meanOfMeasured)**2).sum()
        return 1 - estimateError/variability
    

    
    def LinearRegression(self):
        measuredX = np.array(self.X)  #注意
        measuredX.sort()
        measuredY = self.Y

        i=0
        a=[]
        R=[]
        predictedY = []
        while (i<1):
            a.append([0,0])
            R.append(None)
            predictedY.append(None)
            i=i+1
        
        i=0
        while (i<1):
            a[i][0-i:1] = np.polyfit(self.X, self.Y, i+1)
            predictedY[i] =(a[i][0]*measuredX + a[i][1])
            predictedY1 =(  a[i][0]*self.X + a[i][1])
            R[i] = self.rSquared(measuredY, predictedY1)
            if (R[i]<0)or(R[i]>1):
                predictedY[i] =self.Y  
            i+=1
        return a,measuredX,predictedY,R
    

    
    def export(self,ax,j=5): #参数j在拟合最后一组数据时起作用，调用最后一组数据的export函数，使j=0<5，从而只画linear fit的图
        a,measuredX,predictedY,R = self.LinearRegression()
        
        ax.plot(self.X, self.Y, 'bo')
        for i in range(3):
            if j<5:
                i=j
            if i==0:
                if (R[i]<0)or(R[i]>1):
                    ax.plot(measuredX, predictedY[i], 'g:', label = 'linear fit x='+'%.3f' %(measuredX[0]))
                else :
                    ax.plot(measuredX, predictedY[i], 'g:', label = 'linear fit y='+'%.3f' % (a[i][2])+'*x+'+'%.3f' % (a[i][3]))
            elif i==1:
                ax.plot(measuredX, predictedY[i], 'r:', label = 'square fit y='+'%.3f' % (a[i][1]))
            elif i==2:
                ax.plot(measuredX, predictedY[i], 'y-', label = 'cubic fit y='
                         +'%.3f' % (a[i][0])+'*x^3+'+'%.3f' % (a[i][1])+'*x^2+'+'%.3f' % (a[i][2])+'*x+'+'%.3f' % (a[i][3]))
            if j<5:
                break
                
        ax.set_title('Group'+'%.0f' % (self.j+1))
        ax.legend(loc='best')
        ax.set_xlabel('X')
        ax.set_ylabel('Y')
        return ax
```

### （2）从Excel表格中读取数据

* def read_datafile(filename)：通过引入xlrd模块，读取Excel表格中的数据，每组X和Y数据分别储存到不同的Group类中，最后返回以Group类为元素的group列表。

```python
import xlrd


def read_datafile(filename):
    
    datafile = xlrd.open_workbook(filename)
    sheet = datafile.sheets()[0]
    col = sheet.ncols 
    row = sheet.nrows
    
    group = []
    x=[]
    y=[]
    
    i=0
    while (i < col):
        x.append(None)
        y.append(None)
        group.append(None)
        i=i+1
    
    i=0
    j=0
    while (i < col):
        x[i]=sheet.col_values(i)
        del x[i][0]
        y[i]=sheet.col_values(i+1)
        del y[i][0]
        group[j] = Group(x[i],y[i],j)
        j=j+1
        i=i+2
        
    return row,col,group
```

### （3）main主函数

* 读取表格中相应的数据，并输出结果。

```python
#main

row,col,group = read_datafile('./data.xlsx')
row=row-1
a=int(col/2)
x=[]
y=[]
r=[]
fig = plt.figure(figsize=(15,10)) 
for i in range(a):
    ax = fig.add_subplot(2,2,i+1)
    print (ax)
    group[i].export(ax)
    x.append(None)
    y.append(None)
    r.append(None)
    x[i],y[i],r[i]=group[i].indicators()
    
plt.show()
    
print ('{:>6}'.format('Group'),'{:>10}''{:>10}''{:>10}''{:>10}''{:>10}''{:>10}''{:>10}''{:>10}''{:>10}'
       '{:>10}''{:>10}'.format('x_avg','x_std','x_pstd','x_var','x_pvar','y_avg','y_std','y_pstd','y_var','y_pvar','r'))
for i in range(a):
    print ('%6d'%(float(i+1)),'%*.*f'%(10,3,x[i]['x_avg']), '%*.*f'%(9,3,x[i]['x_std']),'%*.*f'%(9,3,x[i]['x_pstd']),
           '%*.*f'%(9,3,x[i]['x_var']),'%*.*f'%(9,3,x[i]['x_pvar']),'%*.*f'%(9,3,y[i]['y_avg']),'%*.*f'%(9,3,y[i]['y_std']),
           '%*.*f'%(9,3,y[i]['y_pstd']),'%*.*f'%(9,3,y[i]['y_var']),'%*.*f'%(9,3,y[i]['y_pvar']),'%*.*f'%(9,3,r[i]))
```


## 3.分析

* **分析方法及分析数据**

  * 由最后的数据可以看出，四组数据中X和Y的平均值、方差等完全相同，仅从这些分析数据来看，四组数据似乎具有相同的特性，但由散点图可以看出，四组数据并不是在描述同一种关系，所以单从一些分析数据看问题是片面的。

  * 四组数据的相关系数r完全相同，但是由散点图和拟合曲线可以看出，数据分布和拟合曲线完全不同，说明虽然r 的绝对值越大表明相关性越强，但并不存在因果关系。

  * 通过另一个系数$R^2$来衡量，不同的拟合曲线$R^2$不同，越接近1，说明回归方程拟合优度越好，可以看出第二组数据二次多项式拟合时$R^2$极为接近1，所以拟合更好。其中$R^2$的计算公式如下：
$$R^2=\frac{\sum_{i}(y_i-p_i)^2}{\sum_{i}(y_i-\mu)^2}$$

  * 分析数据中百分比也要注意，因为百分比往往会使得人们忽略数据的基数，而导致错误的结果。

  可以看出，不同的分析方法得到不同的分析数据，从不同的分析数据来看待问题会有不同的结果，所以要选择合适的分析方法。


* **描述方法**

  描述数据可以有不同的形式，可以用数字描述，也可以用图像描述，数字描述有时不太能说明数据的关系，而图像描述更为直观，可以选择其中更好的描述方法。图像描述时，图像虽然直观，但不一定可以展现出数据间的真正关系，要注意其横纵坐标的选择、数据的范围等，否则很有可能得出与实际相反的结果。
  
  
* **数据源**
  
  从原始数据看，有时可以发现与其他数据存在明显偏差的数据，这时要思考是否剔除，也许剔除后有更符合实际的结果。比如第三组和第四组数据的个别点。
  
  
* **联系实际**

  我对上述四组数据采取了线性拟合和多项式拟合，可能发现多项式拟合和线性拟合几乎相同，甚至更好，但是要考虑实际来取舍，如果实际问题中X和Y是一次线性关系，即使多项式拟合的效果更好，也应该舍弃。
